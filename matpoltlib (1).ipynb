```python
import pandas as pd
```

```python
df=pd.read_excel('https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/main/Customer%20Call%20List.xlsx')
```

```python
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.drop(columns='Not_Useful_Column',axis=1,inplace=True)
```

```python
df['Last_Name']=df['Last_Name'].str.strip('123./_')
```

```python
df.Last_Name
```

```python
df.loc[:,"Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
```

```python
df.loc[:,'Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
```

```python
df.loc[:,'Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('Na--','')
```

```python
df.Phone_Number
```

```python
df[['street_address','state','zip_code']]=df['Address'].str.split(',',2,expand=True)
```

```python
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('NaN',' ')
```

```python
df
```

```python
df=df.replace('N/a','')
```

```python
df.fillna('',inplace=True)
```

```python
for x in df.index:
  if df.loc[x,'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)
```

```python
for x in df.index:
  if df.loc[x,'Phone_Number'] == '':
    df.drop(x, inplace=True)
```

```python
df=df.reset_index(drop=True)
```
